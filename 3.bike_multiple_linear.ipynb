 Variance Inflation Factor (VIF)
Variance Inflation Factor measures the degree of multicollinearity (i.e., correlation) between predictors. High multicollinearity can lead to instability in coefficient estimates, making the model sensitive to small changes in the data.

Key Points:
Interpreting VIF: A VIF above 5-10 indicates high multicollinearity, suggesting that some predictors may need to be removed or combined.
Impact: High multicollinearity makes it difficult to determine the unique contribution of each predictor, leading to unreliable coefficients.

RFE aims to identify the most significant features for making predictions by recursively removing the least important ones.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

The algorithm starts with all available features.

A model is trained on the dataset, and each features importance is calculated.

After training, RFE ranks features based on their importance to the model.

The least important feature (or a set of features, if specified) is removed from the dataset.

The model is retrained on the remaining features.

The ranking and removal process is repeated until the desired number of features is reached or all features have been removed one-by-one.

RFE stops when it reaches the specified number of features, or all features are ranked.

The remaining features represent the most relevant subset of features for the model.

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=15)  # Specifying n_features_to_select instead of a second argument
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
X_train_rfe
```

# Building Linear Model using 'STATS MODEL'

```python
X_train_rfe.shape[1]
```

```python
X_train_rfe.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['features']=X_train_rfe.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe.values,i)for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
#A VIF above 5-10 indicates high multicollinearity, suggesting that some predictors may need to be removed or combined.
```

```python
import statsmodels.api as sm
# ADD A CONSTANT
X_train_lm1=sm.add_constant(X_train_rfe)
# CREATE FITTED MODEL
lr1=sm.OLS(y_train,X_train_lm1).fit()
```

```python
lr1.params
```

```python
print(lr1.summary())
```

Insights

This model looks good, as there seems to be VERY LOW Multicollinearity between the predictors and the p-values for all the predictors seems to be significant. For now, we will consider this as our final model (unless the Test data metrics are not significantly close to this number).
