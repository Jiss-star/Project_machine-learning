```python
X=df.drop('quality',axis=1)
y=df['quality'].apply(lambda value:1 if value>=7 else 0)# 1 means -good 0 bad
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

#  NORMALIZATION
Normalize X data to decrease effect of large number.
Here, Using MinMaxScaler and ranged in [0, 1]

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
# X=pd.DataFrame(scaler, columns=X_train.columns)
# X.describe().T.style.background_gradient(axis=0, cmap='Purples')
```

```python
X_train
```

```python
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
ypred=classifier.predict(X_test)
```

```python
score=accuracy_score(ypred,y_test)
```
